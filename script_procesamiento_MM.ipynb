```python
import pandas as pd

# Leer el archivo de texto
df_2002 = pd.read_csv('mortalidad_materna_2002.csv', sep=',')  # Cambia el separador si es necesario
df_2002
```

```python
# Agrupar por ENTIDAD_RESIDENCIAD
grouped = df_2002.groupby('ENTIDAD_RESIDENCIAD')

# Iterar sobre cada grupo
for entidad, group in grouped:
    # Ordenar el grupo por CAUSA_CIE_4D
    sorted_group = group.sort_values(by='CAUSA_CIE_4D')
    
    # Guardar el DataFrame ordenado en un archivo CSV
    sorted_group.to_csv(f'2002_{entidad}.csv', index=False)
```

```python
import pandas as pd

# Supongamos que df_2002 ya está definido y contiene las columnas necesarias

# Definir un diccionario para clasificar las causas
causas = {
    'Abortive Outcome': ['O00', 'O01', 'O02', 'O03', 'O04', 'O05', 'O06', 'O07', 'O08'],
    'Sepsis': ['O23', 'O85', 'O86'],
    'Haemorrhage': ['O44', 'O45', 'O67', 'O72'],
    'Hypertension': ['O10', 'O11', 'O12', 'O13', 'O14', 'O15', 'O16'],
    'Dias': ['096'],
    'Total indirect obstetric causes': ['098','099'],
    'Other direct obstetric causes': ['O21', 'O24', 'O25', 'O26', 
                                      'O28', 'O29', 'O30', 'O31',
                                      'O32', 'O33', 'O34', 
                                      'O35', 'O36', 
                                      'O40', 'O41', 
                                      'O42', 'O43',
                                      'O47', 
                                      'O48', 'O60', 'O61', 'O62', 'O63',
                                      'O64', 'O65', 'O66', 'O68', 
                                      'O69', 'O70', 'O71', 'O74', 'O75',
                                      'O81', 'O82', 'O89', 'O90', 'O92',
                                      'O94', 'O95', 'O88', 'O22', 'O87','O73']   
}

# Crear una nueva columna para clasificar las causas
def clasificar_causa(codigo):
    for causa, codigos in causas.items():
        if codigo.startswith(tuple(codigos)):
            return causa
    return None  # Otras causas no especificadas

# Aplicar la función para clasificar las causas
df_2002['Causa Clasificada'] = df_2002['CAUSA_CIE_4D'].apply(clasificar_causa)

# Agrupar por ENTIDAD_RESIDENCIAD y Causa Clasificada
grouped = df_2002.groupby(['ENTIDAD_RESIDENCIAD', 'Causa Clasificada'])

# Iterar sobre cada grupo y guardar en archivos separados
for (entidad, causa), group in grouped:
    # Ordenar el grupo por CAUSA_CIE_4D si es necesario
    sorted_group = group.sort_values(by='CAUSA_CIE_4D')
    
    # Guardar el DataFrame ordenado en un archivo CSV
    filename = f'2002_{entidad}_{causa}.csv'  # Nombre del archivo basado en entidad y causa
    sorted_group.to_csv(filename, index=False)
```
